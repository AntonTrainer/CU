## Q14: (3 Points) Create a new column in `ordL` named, 'revenue'

## Q15-17 code: Create a new dataframe, `custRev`, that only shows the customer name and the sum of the revenue per customer in descending order

```python
names = ordL.groupby(['customername']).revenue.sum()
custRev = pd.DataFrame(names).sort_values('revenue', ascending = False)

custRev.loc['Frau da Collezione']
custRev.loc['Muscle Machine Inc']
custRev.loc['Australian Collectors Co.']
```

## Q18: (3 Points) Calculate the average customer revenue in the `custRev` DataFrame
>- round to two decimal places
>- store the result in a variable called `avgRev`

```python
avgRev = round(custRev.revenue.mean(),2)
avgRev
```

## Q19-21 Code: Show the code to create a new column 'avgdiff' to show the difference in customer revenue and average customer revenue (`avgRev`)

```python
custRev['avgdiff'] = round(custRev.revenue - avgRev,2)
custRev.loc['Mini Gifts Distributors Ltd.']
custRev.loc['La Rochelle Gifts']
```
