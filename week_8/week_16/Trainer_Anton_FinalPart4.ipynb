# Final Part 4 DataFrame and Series Setup
## Import Modules Necessary to Read Data from Yahoo Finance and Work with pandas

```python
import pandas as pd 
import datetime
import pandas_datareader as pdr
```

## Import additional modules we covered in class that allow us to plot graphs

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import style
import matplotlib as mpl
```

## Set up variables to define start and end dates for stock analysis
>- We will look at stock prices over the past 4 years
>- Start date: 2010-05-01
>- End date: 2020-05-01

```python
start = datetime.datetime(2010, 5, 1)
end = datetime.datetime(2020, 5, 1)
```

## Create a DateFrame and/or multiple DataFrames for the following symbols
### Use pandas_datareader and yahoo finance to pull data for:
>- VTI, this is an ETF for the total us stock market
>- BND, an ETF for the total US bond market
>- VEU, an ETF for the total world stock market minus the US
>- BNDX, an ETF for the total world bond market minus the US
>- VNQ, a real estate ETF
>- VDE, an energy sector ETF
>- JETS, an airline sector ETF

#### Note: Having a mixture of these ETFs (especially the first 4 on the list) could be a good way to build a diversified portfolio
>- Not making financial advice here but the point is this is real world stuff that major companies use to build portfolios

```python
vti = round(pdr.DataReader('vti', 'yahoo', start, end),2)
bnd = round(pdr.DataReader('bnd', 'yahoo', start, end),2)
veu = round(pdr.DataReader('veu', 'yahoo', start, end),2)
bndx = round(pdr.DataReader('bndx', 'yahoo', start, end),2)
vnq = round(pdr.DataReader('vnq', 'yahoo', start, end),2)
vde = round(pdr.DataReader('vde', 'yahoo', start, end),2)
jets = round(pdr.DataReader('jets', 'yahoo', start, end),2)
```

## Create Series for Each of the ETFs to store only the `Adj Close` Price
>- This will help you plot graphs individually or with ETFs on the same chart

```python
vti_aclose = vti['Adj Close']
bnd_aclose = bnd['Adj Close']
veu_aclose = veu['Adj Close']
bndx_aclose = bndx['Adj Close']
vnq_aclose = vnq['Adj Close']
vde_aclose = vde['Adj Close']
jets_aclose =jets['Adj Close']
```

# Create Series to Store Daily Returns for Each of the ETFs
>- Create a series that will store daily returns for each ETF
>- Again, this should help you plot returns if/when asked to do so

```python
vti_ret = (vti_aclose/vti_aclose.shift(1))-1
bnd_ret = (bnd_aclose/bnd_aclose.shift(1))-1
veu_ret = (veu_aclose/veu_aclose.shift(1))-1
bndx_ret = (bndx_aclose/bndx_aclose.shift(1))-1
vnq_ret = (vnq_aclose/vnq_aclose.shift(1))-1
vde_ret = (vde_aclose/vde_aclose.shift(1))-1
jets_ret = (jets_aclose/jets_aclose.shift(1))-1
```
