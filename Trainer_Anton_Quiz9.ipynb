# Question 8: 6 points
## Data Preparation - Correctly Filling pandas Null Values
>- Replace the characters you found to be causing pandas to read the data in as objects with null values
>>- Make the change in place for each DataFrame (i.e., don't create new data frames for this)
>>- Lookup the default null value for pandas and use it as your replacement value

```python
# replace the * and # characters with nan values 

#replace all *
bls08.replace('*', np.nan, inplace=True)
bls09.replace('*', np.nan, inplace=True)
bls10.replace('*', np.nan, inplace=True)
bls11.replace('*', np.nan, inplace=True)

# replace all #
bls08.replace('#', np.nan, inplace=True)
bls09.replace('#', np.nan, inplace=True)
bls10.replace('#', np.nan, inplace=True)
bls11.replace('#', np.nan, inplace=True)

bls08.dtypes
```
